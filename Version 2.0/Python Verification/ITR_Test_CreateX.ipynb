### Load CSV Data

```python
df = pd.read_csv('TestSample_100.csv')
df
```

### PreProcess 1: Percentile continuous data to int

```python
q = [10,20,30,40,50,60,70,80,90]
```

```python
c1 = df['Cont1']
perc_c1 = np.percentile(c1, q)
indices_c1 = np.digitize(c1, perc_c1)
indices_c1
```

```python
c2 = df['Cont2']
perc_c2 = np.percentile(c2, q)
indices_c2 = np.digitize(c2, perc_c2)
indices_c2
```

```python
c3 = df['Cont3']
perc_c3 = np.percentile(c3, q)
indices_c3 = np.digitize(c3, perc_c3)
indices_c3
```

```python
df['Cont1'] = indices_c1
df['Cont2'] = indices_c2
df['Cont3'] = indices_c3
df
```
